```python
from PIL import Image, ImageOps
import numpy as np
import matplotlib.pyplot as plt
```

```python

#  & 
img = Image.open('imgs/girl.jpg').convert('L')

# 
plt.subplot(2, 1, 1) # Index
a = np.array(img)
plt.imshow(a)
plt.gray()
plt.axis('off')
plt.show()
```

```python
# 256
plt.subplot(2, 2, 1)
plt.hist(a.flatten(), bins=np.arange(256 )) #256
# 
plt.show()
```

```python
#8
plt.subplot(2,2,1)
plt.hist(a.flatten(), bins=np.arange(8)) #
plt.show()
```

```python
img = Image.open('imgs/overlight.jpg').convert('L')
# 
plt.subplot(2, 1, 1) # Index
a = np.array(img)
plt.imshow(a)
plt.gray()
plt.axis('off')
plt.show()
```

```python
plt.subplot(2, 2, 1) # Index
img = ImageOps.equalize(img);
a = np.array(img)
plt.imshow(img)
plt.gray()
plt.axis('off')
plt.subplot(2, 2, 3) # Index
plt.hist(a.flatten(), bins=np.arange(256)) 
plt.show()
```

```python
images = ['imgs/shimauma.jpg', 'imgs/usagi.jpg', 'imgs/water.jpg', 'imgs/ct.jpg', 'imgs/house.jpg']
for index in range(5):
  img = Image.open(images[index]).convert('L')
  a = np.array(img)
  plt.subplot(2, 5, index+1) 
  plt.imshow(img)
  plt.gray()
  plt.axis('off')
  plt.subplot(2, 5, index+6) # Index
  plt.hist(a.flatten(), bins=np.arange(256)) 

plt.show()
  
```
