```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

#  & 
img = Image.open('girl.jpg').convert('L')

# 
plt.subplot(2, 1, 1) # Index
a = np.array(img)
plt.imshow(a)
plt.gray()
plt.axis('off')

# 
plt.subplot(2, 1, 2)
plt.hist(a.flatten(), bins=np.arange(256 )) #256
# 
plt.show()
```
