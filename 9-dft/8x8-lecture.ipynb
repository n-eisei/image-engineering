```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
#画像表示関数
def  display(data) :
    fig, ax = plt.subplots()
    ax.imshow(data, cmap='binary_r')
    plt.show()    

#DFT関数
def dft(data) :
    #2 次元高速フーリエ変換で周波数領域の情報を取り出す
    f_uv = np.fft.fft2(data)
    
    # 画像の中心に低周波数の成分がくるように並べかえる
    shifted_f_uv = np.fft.fftshift(f_uv)
    
    # パワースペクトルに変換する
    np.seterr(divide = 'ignore')
    magnitude_spectrum2d =  np.log(np.absolute(shifted_f_uv)) 
    magnitude_spectrum2d *= (255.0/magnitude_spectrum2d.max())
    return magnitude_spectrum2d

#DFT2関数
def dft2(data) :
    #2 次元高速フーリエ変換で周波数領域の情報を取り出す
    f_uv = np.fft.fft2(data)
    
    # 画像の中心に低周波数の成分がくるように並べかえる
    shifted_f_uv = np.fft.fftshift(f_uv)
    
    # パワースペクトルに変換する
    np.seterr(divide = 'ignore')
    magnitude_spectrum2d =  np.absolute(shifted_f_uv)
    magnitude_spectrum2d *= (255.0/magnitude_spectrum2d.max())
    return magnitude_spectrum2d
```

```python
data = np.zeros([8,8])
data[4][4] = 1
#画像表示
display(data)
```

```python
display(dft(data))
```
